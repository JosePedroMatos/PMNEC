### PMNEC

This is a jupyter notebook.  
It is not the best tool to code complex projects, but it is excellent to test and learn.

The first thing to have in mind is that a python environment must be chosen to run the notebook.  
Check the file pmnec.yml. It contains a list of packages necessary to create a `pnmec` conda environment that we will use.  
A conda environment is a "separate" python installation that does not interfere with your system's main python or other environemnts.

Opening `anaconda prompt` and navigating to the folder of this project, you just need to run:
- conda install -n base conda-libmamba-solver
- conda config --set solver libmamba
- conda env create -f pmnec.yml

Once completed, you will be able to choose `pmnec` as your jupyter kernel at the top right of this window.
